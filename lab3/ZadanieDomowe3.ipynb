```r
data = read.csv("results.csv")
```

```r
data
```

Zad 1. Umieszczenie danych w jednej tabeli.


```r
dataChanged <- data %>% 
    group_by(Type, Size) %>%
    summarise(
        sdTime = sd(Time),
        averageTime = mean(Time)
    ) 
dataChanged
```

```r
onlyDots = dataChanged %>% filter (Type == 'dotOperation')
onlyDots
```

```r
onlyMatrix = dataChanged %>% filter (Type == 'matrixOperation')
onlyMatrix
```

Zad 2. Wykresy.

```r
library(ggplot2)
options(repr.plot.width=10, repr.plot.height=5)

dotPlot <- ggplot(onlyDots, aes(x=Size)) + geom_point(aes(y=averageTime), size = 3, color = "red") + 
theme(text = element_text(size = 15)) +
ggtitle("Time of computing dot operation") + xlab("Size") + ylab("Time [s]")
dotPlot
```

```r
library(ggplot2)
options(repr.plot.width=10, repr.plot.height=5)

matrixPlot <- ggplot(onlyMatrix, aes(x=Size)) + geom_point(aes(y=averageTime), size = 3, color = "blue") + 
theme(text = element_text(size = 15)) +
ggtitle("Time of computing matrix operation") + xlab("Size") + ylab("Time [s]")
matrixPlot
```

Zad 3. Dodanie odchylenia standardowego.

```r
dotPlot + geom_errorbar(aes(ymin = averageTime - sdTime, ymax = averageTime + sdTime))
```

```r
matrixPlot + geom_errorbar(aes(ymin = averageTime - sdTime, ymax = averageTime + sdTime))
```
