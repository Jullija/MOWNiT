#### Zad 1. Trzy metody poszukiwania pierwiastków.

a) Metoda wykorzystująca przedział i zmianę czasu - METODA BISEKCJI.

b) Metoda wykorzystująca pochodną - METODA NEWTONA.


c) Metoda wykorzystująca przybliżenie pochodnej - METODA STEFFENSENA.

```julia
using DataFrames, Roots, ForwardDiff
df = DataFrame(fun=Function[], method=String[], iterations=Int64[], funCalls=Int64[])
fun1(x) = sin(x) - x/2
fun2(x) = cos(x) - x
fun3(x) =  20.0 * x / ( 100.0 * x * x + 1.0 )
fun4(x) = 2*x-exp(-x)
fun5(x) = x*exp(-x)
fun6(x) = (x+3)*(x-1)^2
functions=Function[]
push!(functions, fun1)
push!(functions, fun2)
push!(functions, fun3)
push!(functions, fun4)
push!(functions, fun5)
push!(functions, fun5)
push!(functions, fun6)


D(f) = x->ForwardDiff.derivative(f, float(x))

for fun in functions
    tracker = Roots.Tracks()
    bis = find_zero(fun, (0,1), Bisection(), verbose=true, tracks = tracker)
    if !iszero(fun(bis))
        println("Error, " ,fun, " at ", fun(bis))
    end
    
    push!(df, (fun, "Bisection", tracker.steps, tracker.fncalls))
    
    tracker = Roots.Tracks()
    newton = find_zero((fun, D(fun)), 0, Roots.Newton(), verbose=true, tracks = tracker)
    if !iszero(fun(newton))
        println("Error, " ,fun, " at ", fun(newton))
    end
    
    push!(df, (fun, "Newton", tracker.steps, tracker.fncalls))
    
    tracker = Roots.Tracks()
    steff = find_zero(fun, 0, Order2(), verbose=true, tracks = tracker)
    if !iszero(fun(steff))
        println("Error, " ,fun, " at ", fun(steff))
    end

    push!(df, (fun, "Steffensen", tracker.steps, tracker.fncalls))
end

stats = combine(groupby(df, [:fun, :method]), "iterations", "funCalls")
stats
```

#### Zad 2. Wstęga Newtona
Wstęga Newtona to fraktal, który powstaje w wyniku zastosowania metody Newtona i kolorowaniu kolejnych wyników iteracji.

Dzielimy płaszczyznę zespoloną na siatkę kwadratów o równych bokach i stosujemy metodę Newtona dla każdego punktu w środku każdego kwadratu. Dla każdego punktu, iterujemy wzór $z=z−\frac{f(z)}{f′(z)}$ aż do znalezienia pierwiastka lub przekroczenia maksymalnej liczby iteracji.

Kolorujemy dany punkt na płaszczyźnie zespolonej na określony kolor odpowiadający pierwiastkowi. Punkty, które nie dążą do żadnego pierwiastka, są kolorowane na czarno lub w inny sposób oznaczane.

```julia
using Statistics, Measures, Symbolics, Plots

f(x) = x^5 - 1
fDeriv(x) = 5*x^4
n = 100

function z(x)
    return x - f(x) / fDeriv(x)
end

function iterationUpdate(x, update)
    for i in 1:n
        x = update(x)
    end
    return x
end
update = eval(z)
lower = -2 - 2im
upper = 2 + 2im
step = 0.5e-2
Z0 = [a+b*im for b in real(lower):step:real(upper),
                    a in imag(lower):step:imag(upper)]
Z1 = iterationUpdate.(Z0, update)
heatmap(angle.(Z1), colorbar=false, color=:rainbow, ticks=false)
```
