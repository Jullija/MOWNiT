```julia
# get polynomial array
p0 = Polynomial([1.0])
p1 = Polynomial([0.0,1.0])
ps = [p0, p1]
for k in 1:6
    p = ((2*k+1)/(k+1))*p1*ps[k+1] - (k/(k+1))*ps[k]
    push!(ps, p)
end

# print polynomials
xs = -1:1
plot(p0,extrema(xs)...)
for i in 2:6
    plot!(ps[i],extrema(xs)...)
end
display(plot!())
```

### Zadanie 2 (1p)
- Napisz funkcję, która dla podanej liczby punktów Gaussa k oraz funkcji f policzy $\int_{-1}^{1} fdx$ metodą używającą funkcji: gauss(k) oraz sumy sum(a .* f.(xp))
- przetestuj dla wielomianów coraz większych stopni
- sprawdz kiedy przestaje być dokładna,
- podaj związek z twierdzeniem o stopniu dokładności kwadratury Gaussa

```julia
function integral_Gauss(n,f)
    (roots, coeffs) = gauss(n)
    return sum(coeffs.*f.(roots))
end
```

```julia
test = [[if j != i 0 else i end for i in 1:j] for j in 1:10]
```

```julia
degree = 1

for poly in test
    pn = Polynomial(poly)
    println()
    println("degree: ", degree)
    println(pn, "     ", integral_Gauss(degree, pn))
    println("expected: ", 1^length(poly) - ((-1)^length(poly)))
    degree += 1
end
```

### Zadanie 3 (1p)
Skorzystaj z rozwiązania zadania 2 do napisania funkcji liczącej całki w dowolnym przedziale 
$\int_{a}^{b}f(x)dx$
dokonując normalizacji do: 
$\int_{-1}^{1}F(z)dz$
podstawiając:

$$x=\frac{b+a}{2}+\frac{b-a}{2}$$
$$dx=\frac{b-a}{2}dz$$

Przetestuj działanie na kilku przykładach i sprawdź z wynikami otrzymanymi analitycznie.

```julia
function integral_Gauss_interval(n, f, a, b)
    x, w = gauss(n)
    x = (b-a)/2 *x .+ (b+a)/2
    res = (b-a)/2 * sum(f.(x) .* w)
    return res
end
```

```julia
println("expected value: ", 80)
f = x -> 4*x^3
for i in 1:5
    println("i: ", i," res: ", integral_Gauss_interval(i, f, 1, 3))
end
```

```julia
println("expected value: ", 2)
f = sin
for i in 1:5
    println("i: ", i," res: ", integral_Gauss_interval(i, sin, 0, pi))
end
```

```julia
println("expected value: ", exp(1))
f = x -> exp(x)
for i in 1:5
    println("i: ", i," res: ", integral_Gauss_interval(i, f, 0, 1))
end
```

### Zadanie 5 (3p)
Napisz własne funkcje całkujące - w wybranym przedziale (a,b) - złożonymi metodami prostokątów, trapezów oraz Simpsona. Dla stałego przedziału zwiększaj liczbę użytych punktów (n). Narysuj wykresy funkcji błędu w stosunku do wyniku otrzymanego analitycznie, w zależnosci od 
$h=\frac{b-1}{n}$:
- wykładniczej
- trygonometrycznej (np. sin (x) w przedziale [0,$2pi$])

```julia
function rectangle(n, f, a, b)
    sum = 0
    h = (b-a)/n
    for i in 1:n
       sum += f(a + (i-1)*h + h/2)*h
    end
    return sum
end
```

```julia
function trapeze(n, f, a, b)
    sum = 0
    h = (b-a)/n
    for i in 1:n
        a1 = f(a+(i-1)*h)
        a2 = f(a+(i*h))
        sum += 0.5*(a1+a2)*h
    end
    return sum
end
```

```julia
function Simpson(n, f, a, b)
    dx = (b-a)/n
    st = 0
    s = 0
    for i in 1:n
        x = a + i * dx;
        st += f(x-dx/2);
        if(i < n) 
            s += f(x)
        end
    end
    return dx/6*(f(a) + f(b) + 2*s + 4*st);
end
```

```julia
using DataFrames
expected = 43.281
a = 1
b = 5
testFun(x) = 2^x
df = DataFrame(Points=Int64[], h = Float64[], Type=String[], Error=Float64[])
for i in 10:100
    h = (b-a)/i
    push!(df, (i, h, "rectangle", abs(expected - rectangle(i, testFun, a, b))))
    push!(df, (i, h, "trapez", abs(expected - trapeze(i, testFun, a, b))))
    push!(df, (i, h, "Simpson", abs(expected - Simpson(i, testFun, a, b))))
end
stats = combine(groupby(df, [:Type, :Points, :h]), "Error")
stats
```
